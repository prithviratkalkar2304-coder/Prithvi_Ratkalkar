#Study of Sets in Python

```python
#Creation of Set

thisset={"Raj", "Adi","Yash"}
print(thisset)   #(Unordered indexing)

```

```python
#Duplicates are not allowed
thisset={"Raj","Raj", "Adi","Yash","Raj"}
print(thisset)
```

```python
#the value True and 1 are considered same value in sets
thisset={"Raj","1","Raj", "Adi","Yash","1","Raj",True,1,0 ,False}
print(thisset)
```

```python
#A set with string, integer, boolean values
set1={"Aditya", 5,True, 14.34}
print(set1)
```

```python
#searching
thisset={"Raj", "Adi","Yash"}
print("Adi" in thisset)
print("Adi" not in thisset)
print("Aditya" in thisset)
```

```python
# once a set is created you can't replace its items but you can add new items
thisset={"Raj", "Adi","Yash"}
thisset.add("Aditya")
print(thisset)
```

```python
# remove items
thisset={"Raj", "Adi","Yash"}
thisset.remove("Adi")
print(thisset)
```

```python
#Set operators
A={1,2,3,4,5}
B={4,5,6,7,8}
print("Union", A|B)
print("Intersection", A&B)
print("Difference", B-A)
print("Symmetric Difference", A^B)
```

```python
#Problem-2
A={"Maths","Chemistry","Physics","Computers","Literature","German"}
B={"English", "Geography", "Economics", "Maths"}
C={"Maths","Geography","History" }
print("The common subject between A and B are:", A&B)
print("The common subject between B and C are:", B&C)
print("The common subject between A, B, C are:", A&B&C)
```

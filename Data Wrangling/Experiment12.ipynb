```python
import pandas as pd
data={
    "Roll No":[101,102,103,104,105],
    "Name":["Prithvi","Akshay","Raj","Diya","Sam"],
    "Age":[20,"-",21,23,"18"],
    "Marks":[89,87,90,'-',95],
    "Department":["CSE","ENTC","it","MECH","cse"],
    "Admission_Date":["01-06-2025","23-03-2025","07-07-2025","14-06-2025","26/10/2025"]
}
df=pd.DataFrame(data)
df
```

```python
#Step 2:Replace '-'with NaN to mark the missing values
df.replace("-",np.nan,inplace=True)
#Comvert columns to numeric where needed
df["Age"]=pd.to_numeric(df["Age"],errors='coerce')
df["Marks"]=pd.to_numeric(df["Marks"],errors='coerce')

```

```python
#Step3: Check the missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
#Step4: Fill missing Age with mean
df["Age"].fillna(df["Age"].mean(),inplace=True)
#Fill missing Marks with median
df["Marks"].fillna(df["Marks"].median(),inplace=True)
df
```

```python
#Step5: Fix the inconsistent text format
df["Department"]=df["Department"].str.upper()
df
```

```python
#Step6: Convert Admission_Date to standard datetime format
df["Admission_Date"]=pd.to_datetime(df["Admission_Date"],errors='coerce')
df
```

```python
# Step7: Print clean data
print("\nCleanedData:")
df
```

```python
#Step8: Save the cleaned data
df.to_csv("cleaned_student_data.csv",index=False)
print("\nCleaned CSV file saved as cleaned_student_data.csv")
```

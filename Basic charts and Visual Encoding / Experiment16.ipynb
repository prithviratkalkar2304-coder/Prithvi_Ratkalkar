```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data={
    'Days':[ 'Monday','Tuesday','Wednesday','Thursday','Friday'],
    'Study_Hours':[1,10,3,3,8],
    'Marks':[80,70,20,50,100],
    'Attendence':[100,90,80,70,60],
    'Sleep_Hours':[7,7,5,6,9],
    'Assignment_Completed':[1,2,1,3,2]
}
df=pd.DataFrame(data)
df
```

```python
#1 Line Chart(Trend)
plt.plot(df['Days'],df['Study_Hours'],marker='*')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Study Hours Over Days')
plt.show()
```

```python
#Advanced Line Chart
plt.figure(figsize=(7,4))
plt.plot(df['Days'],df['Study_Hours'],marker='o',color='red',label='Study Hours')
plt.plot(df['Days'],df['Marks'],marker='^',color='green',label='Marks')
plt.title('Study Hours and Marks Over Days')
plt.xlabel('Days')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
```

```python
#2 ar Chart (Comparision)
plt.bar(df['Days'],df['Marks'],color='pink')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.title('Marks Over Days')
plt.show()
```

```python
#Advanced Bar Graph
bars=plt.bar(df['Days'],df['Marks'],color='teal')
for bar in bars:
  y=bar.get_height()
  plt.text(bar.get_x() + bar.get_width()/2,y,str(y),ha='center',va='bottom')
plt.title('Marks Comparision')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.grid(axis='y')
plt.show()
```

```python
#3 Histogram
data={
    'Marks':[50,60,55,70,65,75,80,85,90,45,69,40,74,83,70]
}
df=pd.DataFrame(data)
plt.hist(df['Marks'],bins=5,color='black',edgecolor='black',alpha=0.75)
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Marks Distribution')
```

```python
#4 Scatter Plot(Relationship)
import matplotlib.pyplot as plt
plt.scatter(df['Study_Hours'],df['Marks'])
plt.xlabel('Study_Hours')
plt.ylabel('Marks')
plt.title('Marks vs Study_Hours')
plt.show()
```

```python
# Coditional Scatter Plot
df['Result']=['Fail','Pass','Pass','Fail','Pass']
colors=['red' if r=='Fail' else 'green' for r in df['Result']]
plt.scatter(df['Study_Hours'],df['Marks'],c=colors)
```

```python
# 2nd Data set
data ={
    'Day':['Monday','Tuesday','Wednesday','Thursday','Friday'],
    'Region':['North','South','East','West','North'],
    'Sales':[200,220,250,270,300],
    'Profit':[100,120,150,170,200],
    'Customers':[1000,1200,1500,1700,2000],
    'Category':['Electronics','Clothing','Electronics','Clothing','Electronics']
}
df=pd.DataFrame(data)
df
```

```python
#Line Plot
sns.lineplot(x='Day',y='Sales',data=df)
plt.xlabel('Day')
plt.ylabel
```

```python
# Bar Graph
sns.barplot(x='Region',y='Sales',data=df)
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

```python
#Histogram
sns.histplot(df['Sales'],bins=3,color='red')
plt.xlabel('Sales')
```

```python
#Scatter Plot
sns.scatterplot(x='Customers',y='Profit',data=df)
plt.xlabel('Customers')
```

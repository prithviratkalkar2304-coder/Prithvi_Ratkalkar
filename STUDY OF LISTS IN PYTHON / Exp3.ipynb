```python
#Extend list
thislist=["Cricket","Football","Hockey"]
this=['A',3,True]
thislist.extend(this)
print(thislist)
```

```python
#Problems-2
'''2. Grocery Shopping List
A user maintains a grocery list for monthly shopping
Tasks:
* Create a grocery list
* Add new items to list
* Remove purchased items
*display the final shopping list
Consepts Used: append(),remove(),indexing'''


GL1=["peanut butter","eggs","vegies","apples"]
GL2=["cheese","banana","bread"]
GL1.append("milk")
print(GL1)
GL1.remove("peanut butter")
print(GL1)
GL1.extend(GL2)
print(GL1)
```

```python
#Problem-3
'''3. A person stores phone contacts in a list.
Tasks:
*Display the list
*Add a new Contact
*Remove a contact
*Arrange the contacts in alphabetical order
Concepts used: sort()...
'''
c=["Prithvi", "Aaryan", "Astha", "Aditya", "Raj", "Yash"]
print(c)
c.append("Aryan")
print(c)
c.remove("Raj")
print(c)
print(sorted(c))
```

```python
#Problems-4
''' 4. Temperature Analysis
Daily temperature readings of a city are stored in a list for 10 days
* Display temperatures of first and last 5days
* Find highest and lowest temp
* Calculate average temp
Concepts used: max(),min(),sum()'''
Pune=[12,23,27,30,28,28,32,25,24,20]
print(Pune[:5])
print(Pune[-5:])
print(max(Pune))
print(min(Pune))
print(sum(Pune)/len(Pune))
```

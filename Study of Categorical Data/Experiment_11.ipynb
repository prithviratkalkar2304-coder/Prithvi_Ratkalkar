```python
import pandas as pd
data = {
    "Order_ID":['01','02','03','04','05','06','07','08','09','010'],
    "Category":['Electronics','Clothing','Electronics','Grocery','Clothing','Electronics','Grocery','Clothing','Electronics','Grocery'],
    "Payment_Method":['Credit Card','Cash','Debit Card','Credit Card','Cash','Debit Card','Credit Card','Cash','Debit Card','Credit Card'],
    "Delivery_Type":['Standard','Express','Standard','Express','Standard','Express','Standard','Express','Standard','Express'],
    "Customer_Type":['Regular','Occasional','Regular','Occasional','Regular','Occasional','Regular','Occasional','Regular','Occasional']
}

df=pd.DataFrame(data)
df.to_csv("Shop.csv",index=False)
```

```python
#Frequency Count
```

```python
#2 unque values
print(df['Category'].unique())
```

```python
#4
print(df['Payment_Method'].value_counts())
```

```python
#5 Cross Tabulation
#Category vs Payment
print(pd.crosstab(df['Category'],df['Payment_Method']))
```

```python
#Delivery vs payment
print(pd.crosstab(df['Delivery_Type'],df['Payment_Method']))
```

```python
#Percentage distribtion
print(df['Category'].value_counts(normalize=True)*100)
```

```python
# Filtering Data
#Eg: Orders from Electronics category
print(df[df['Category']=='Electronics'])
```

```python
#Gruping
# Orders by category and payment method
print(df.groupby(['Category','Payment_Method']).value_counts())
```

```python
# Sorting Data
sorted_df = df.sort_values(by='Category')
print(sorted_df)
```
